```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n=200
hours_slept=np.random.normal(7, 1.5, size=n)

cups_of_coffee=5-0.5*hours_slept+np.random.normal(0, 0.5, size=n)
productivity=3+0.8*hours_slept+np.random.normal(0,1, size=n)
stress=8-0.6*hours_slept+np.random.normal(0,1,size=n)

df=pd.DataFrame({
    "Hours_slept": hours_slept,
    "Cups_of_coffee": cups_of_coffee,
    "Productivity": productivity,
    "Stress": stress
})
print(df.head())
print(df.describe())

cov_matrix=df.cov()
print(cov_matrix)
    
```

```python
corr_matrix=df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Layer 2: Correlation Matrix Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist2d(df["Hours_slept"], df["Cups_of_coffee"], bins=20, cmap="YlOrRd")
plt.colorbar(label="Frequency")
plt.xlabel("Hours Slept")
plt.ylabel("Cups of Coffee")
plt.title("Layer 3: Joint Probability Density--Hourse Slept vs Coffee")
plt.show()
```
