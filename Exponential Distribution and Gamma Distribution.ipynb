```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lam=10
scale=60/lam
n_simulations=1000

inter_arrival_times=np.random.exponential(scale=scale, size=n_simulations)

plt.hist(inter_arrival_times, bins=30, edgecolor="black")
plt.xlabel("Time Between Arrivals (minutes)")
plt.ylabel("Frequency")
plt.title("Layer1: Exponential Inter-Arrival times")
plt.show()
```

```python
k=10
n_simulations=5000

gamma_samples=np.array([np.sum(np.random.exponential(scale=scale, size=k))
                        for _ in range(n_simulations)])

plt.hist(gamma_samples, bins=50, edgecolor="black")
plt.xlabel("Total wait for 10th Customer( minutes)")
plt.ylabel("Frequency")
plt.title("Layer 2: Gamma Distribution (Sum of 10 Exponentials)")
plt.show()
```

```python
from scipy import stats

fig, axes=plt.subplots(1,2, figsize=(14, 5))

axes[0].hist(inter_arrival_times, bins=30, edgecolor="black", color="steelblue")
axes[0].set_title("Layer1 :Exponential (Single Arrival Gap)")
axes[0].set_xlabel("Time Between Arrivals(minutes)")
axes[0].set_ylabel("Frequency")

axes[1].hist(gamma_samples, bins=50, edgecolor="black", color="steelblue", density=True)
x=np.linspace(0, 200, 1000)
pdf=stats.gamma.pdf(x, a=10, scale=scale)
axes[1].plot(x, pdf, color="red", linewidth=2, label="Therotical Gamma PDF")
axes[1].set_title("Layer2: Gamma (Wailt for 10th customer)")
axes[1].set_xlabel("Total Wait (minutes)")
axes[1].set_ylabel("Density")
axes[1].legend()

plt.tight_layout()
plt.show()
```
