```python
import numpy as np
import matplotlib.pyplot as plt

p=0.55
n_flips=10000

flips=np.random.binomial(1, 0.55, size=n_flips)
running_avg=np.cumsum(flips)/np.arange(1, n_flips+1)

plt.plot(running_avg, color="steelblue")
plt.axhline(y=0.55, color="red", linestyle="--", label="True probability (0.55)")
plt.xlabel("Number of Flips")
plt.ylabel("Running Average")
plt.title("Layer 1: Law of Large Numbers- The House Always Wins")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n_samples=1000
sample_size=30

sample_means=np.array([np.mean(np.random.uniform(0, 1, size=sample_size))
                       for _ in range(n_samples)])

fig, axes=plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(np.random.uniform(0,1, size=10000), bins=50, edgecolor="black", color="steelblue")
axes[0].set_title("Original: Uniform Distribution")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("frequency")

axes[1].hist(sample_means, bins=50, edgecolor="black", color="steelblue")
axes[1].set_title("Clt: Distribution of Sample Means")
axes[1].set_xlabel("Sample Mean")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
