```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
df_results=pd.read_csv("investment_parameters.csv")
budget=pd.read_csv("budget_clean.csv")
```

```python
top10=budget.head(10)

plt.figure(figsize=(12, 6))
plt.barh(top10["Ministry"], top10["Total_Allocation"], color="steelblue")
plt.xlabel("Allocation (₹ Crores)")
plt.title("Top 10 Ministries by Budget Allocation 2026-27")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("chart1_budget.png", dpi=150)
plt.show()
```

```python
colors={"Infrastructure": "steelblue", "DEFENCE": "crimson", "AGRICULTURE": "green"}

plt.figure(figsize=(12, 6))
for sector in df_results["Sector"].unique():
    sector_data=df_results[df_results["Sector"]==sector]
    plt.bar(sector_data["Stock"], sector_data["Cumulative_Return"],
            color=colors[sector], label=sector)

plt.axhline(y=0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel("Stock")
plt.ylabel("Cumulative Return (%)")
plt.title("1 Year Stock Performance by Sector")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("chart2_returns.png", dpi=150)
plt.show()
```

```python
df_sorted_sharpe=df_results.sort_values("Sharpe_Ratio", ascending=True)

colors_sharpe=df_sorted_sharpe["Sharpe_Ratio"].apply(
    lambda x: "green" if x > 0 else "crimson"
)

plt.figure(figsize=(12, 6))
plt.barh(df_sorted_sharpe["Stock"], df_sorted_sharpe["Sharpe_Ratio"], color=colors_sharpe)
plt.axvline(x=0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel("Sharpe Ratio")
plt.title("Sharpe Ratio Ranking-Best Risk Adjusted Returns")
plt.tight_layout()
plt.savefig("chart3_sharpe.png", dpi=150)
plt.show()
```
