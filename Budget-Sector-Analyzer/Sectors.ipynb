```python
import yfinance as yf
import pandas as pd
```

```python
sectors={
    "Infrastructure":["LT.NS", "IRB.NS", "NTPC.NS", "POWERGRID.NS", "CUMMINSIND.NS"],
    "DEFENCE":["HAL.NS", "BEL.NS", "MIDHANI.NS", "PARAS.NS", "COCHINSHIP.NS"],
    "AGRICULTURE":["UPL.NS", "COROMANDEL.NS", "PIIND.NS", "RALLIS.NS", "CHAMBLFERT.NS"]
}
```

```python
all_data={}

for sector, stocks in sectors.items():
    sector_data=yf.download(stocks, period="1y", auto_adjust=True)["Close"]
    all_data[sector]=sector_data
    print(f"{sector} downloaded successfully")
```

```python
print(all_data["Infrastructure"].head())
print(all_data["Infrastructure"].shape)
```

```python
returns={}

for sector, data in all_data.items():
    returns[sector]=data.pct_change().dropna()
    print(f"{sector} returns calculated - shape: {returns[sector].shape}")
```

```python
risk_free_rate=0.065
results=[]

for sector, ret in returns.items():
    for stock in ret.columns:
        cum_return=(1+ret[stock]).prod()-1
        volatility=ret[stock].std()*(252**0.5)
        sharpe=(ret[stock].mean()*252-risk_free_rate)/volatility
        
        results.append({ 
            "Sector": sector,
            "Stock": stock,
            "Cumulative_Return": round(cum_return *100, 2),
            "Volatility": round(volatility*100, 2),
            "Sharpe_Ratio": round(sharpe, 2)
        })

df_results=pd.DataFrame(results)
print(df_results)
        
```

```python
df_results.to_csv("investment_parameters.csv", index=False)
print("Saved!")
```
