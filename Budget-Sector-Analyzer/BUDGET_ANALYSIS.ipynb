```python
import pandas as pd
```

```python
df=pd.read_excel("allsbe.xlsx", skiprows=8)
df.head(20)
```

```python
print(df.columns.tolist())
df.iloc[:5, :10]
```

```python
df_clean=df[["Unnamed: 2", "Unnamed: 9"]]
df_clean.columns=["Ministry", "Total_Allocation"]
df_clean.head(20)
```

```python
df_clean=df_clean.dropna(subset=["Ministry", "Total_Allocation"])
df_clean.head(20)
df_clean=df_clean[~df_clean["Ministry"].str.match(r"v\d")]
df_clean.reset_index(drop=True, inplace=True)
df_clean.head(20)

```

```python
df_clean["Ministry"].iloc[1:5].tolist()
```

```python
df_clean=df_clean[df_clean["Ministry"].str.strip().str.len()>4]
df_clean.reset_index(drop=True, inplace=True)
df_clean.head(20)
```

```python
print(df_clean.shape)
```

```python
df_sorted=df_clean.sort_values("Total_Allocation", ascending=False)
df_sorted.head(10)
```

```python
df_sorted.to_csv("budget_clean.csv", index=False)
```
